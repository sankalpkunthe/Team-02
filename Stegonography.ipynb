```python
from PIL import Image
```

```python
img = Image.open("lion.png")
img_encoded = img.copy()
width, height = img.size
```

```python
message = "A paragraph is a self-contained unit of discourse in writing that deals with a single, specific idea or theme. It usually consists of several sentences and is the fundamental building block of essays, articles, books, and reports. The structure of a well-written paragraph typically follows three parts: a topic sentence (which states the main idea), supporting sentences (which provide details, examples, and evidence), and a concluding or transitional sentence (which summarizes the point and often leads to the next paragraph). All sentences within a paragraph must be focused, coherent, and flow logically to develop the central topic." #random text which I took from internet
message_bits = ''.join([format(ord(c), '08b') for c in message])
print(message_bits)
print(len(message_bits))
```

```python
pixels = list(img_encoded.getdata())
num_pixels = len(pixels)
```

```python
def encode_message_in_red(pixels, message_bits):
    encoded_pixels = []
    bit_idx = 0
    for i, (r, g, b) in enumerate(pixels):
        if bit_idx < len(message_bits):
            # Clear last bit of red channel
            r = r & 0b11111110
            # Set last last to secret bits
            r = r | int(message_bits[bit_idx])
            bit_idx += 1
        encoded_pixels.append((r, g, b))
    return encoded_pixels
```

```python
encoded_pixels = encode_message_in_red(pixels, message_bits)
img_encoded.putdata(encoded_pixels)
img_encoded.save("encoded_lion.png")
print("Message encoded successfully!")
```

```python
img_encoded = Image.open("encoded_lion.png")
pixels = list(img_encoded.getdata())
```

```python
def decode_message_from_red(pixels, num_chars):
    bits = ""
    for i in range(num_chars * 8):  # 2 bits per pixel
        r = pixels[i][0]
        # Extract last 2 bits of red channel
        last_bit = r & 0b00000001
        bits += str(last_bit)
         # Convert bits to characters
    message = ""
    for i in range(0, len(bits), 8):
        byte = bits[i:i+8]
        message += chr(int(byte, 2))
    return message

```

```python
decoded_message = decode_message_from_red(pixels, num_chars=len(message))
print("Decoded message:", decoded_message)
```
